```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data file
face_data = np.load('olivetti_faces.npy')
```

```python
# load target file
face_target = np.load('olivetti_faces_target.npy')
```

```python
print(face_data.shape)
```

```python
print(face_target.shape)
print(face_target)
```

```python
# show the first 10 people's 10 images
fig, axes = plt.subplots(10, 10, figsize=(10, 10))
for i in range(10):
    for j in range(10):
        axes[i, j].imshow(face_data[i*10+j], cmap='gray')
        axes[i, j].axis('off')
plt.show()
```

```python
# index list of each person
people_data_index_list = []
for i in range(40):
    people_data_index_list.append(np.where(face_target == i)[0])
people_data_index_list = np.array(people_data_index_list)
print(people_data_index_list)
```

```python
# split train and test index
train_index = people_data_index_list[:,:8].reshape(-1)
test_index = people_data_index_list[:,8:].reshape(-1)
```

```python
# split train and test data and target
train_data = face_data[train_index, :, :]
test_data = face_data[test_index, :, :]
train_target = face_target[train_index]
test_target = face_target[test_index]
```

```python
# show train data
fig, axes = plt.subplots(10, 8, figsize=(9, 10))
for i in range(10):
    for j in range(8):
        axes[i, j].imshow(train_data[i*8+j], cmap='gray')
        axes[i, j].axis('off')
plt.suptitle('Train Data')
plt.show()
```

```python
# show test data
fig, axes = plt.subplots(10, 2, figsize=(3, 10))
for i in range(10):
    for j in range(2):
        axes[i, j].imshow(test_data[i*2+j], cmap='gray')
        axes[i, j].axis('off')
plt.suptitle('Test Data')
plt.show()
```

```python
# make 3d array into 2d array, where each row represents a 64*64 image
train_data = train_data.reshape(train_data.shape[0], -1)
test_data = test_data.reshape(test_data.shape[0], -1)
```

```python
# save train and test data  csv file
np.savetxt('train_data.csv', train_data, delimiter=',')
np.savetxt('test_data.csv', test_data, delimiter=',')
np.savetxt('train_target.csv', train_target, delimiter=',', fmt='%d')
np.savetxt('test_target.csv', test_target, delimiter=',', fmt='%d')
```
